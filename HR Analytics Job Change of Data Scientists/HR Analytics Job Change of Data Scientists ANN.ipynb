```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train,X_val,y_train,y_val=train_test_split(X_test,y_test,test_size=0.2,random_state=0)
```

```python
model=Sequential()
model.add(tf.keras.Input(shape=(X_train.shape[1],)))
model.add(Dense(units=480,kernel_initializer='GlorotUniform',activation='sigmoid'))
model.add(Dense(units=192,kernel_initializer='GlorotUniform',activation='sigmoid'))
model.add(Dense(units=32,kernel_initializer='GlorotUniform',activation='sigmoid'))
model.add(Dense(units=1,kernel_initializer='GlorotUniform',activation='sigmoid'))
opt=Adam(learning_rate=0.001)
bce = tf.keras.losses.BinaryCrossentropy()
model.compile(optimizer=opt,loss=bce)
```

```python
model.summary()
```

```python
model_trained=model.fit(X_train,y_train,batch_size=50,epochs=500,class_weight={0:1,1:9},validation_data=(X_val,y_val))
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Classification Report \n",(classification_report(y_test, (y_pred>0.9))))
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch

def build_model(hp):
    model = keras.Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(layers.Dense(units=hp.Int('units_' + str(i),
                                            min_value=32,
                                            max_value=512,
                                            step=32),
                               activation='sigmoid'))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='categorical_crossentropy',
        metrics='accuracy')
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3)
```

```python
tuner.search_space_summary()
```

```python
tuner.search(X_train,y_train,validation_data=(X_test,y_test))
```
